```python
accountBalance = 1000
pin = "7634"

def PinEntry():
  global pin
  while True:
    print(" ===Enter PIN=== ")

    pinInput = input()

    if pinInput == pin:
      ATMMenu()

    else:
      print("Incorrect pin")

def ATMMenu():
  while True:
    print("==== ATM ====")
    print("1. Check balance")
    print("2. Withdraw money")
    print("3. Deposit money")
    print("4. exit")

    selection = input("Select an option from 1 to 4: ")

    if selection == "1":
      Balance()

    elif selection == "3":
      Deposit()

    elif selection == "2":
      Withdraw()

    elif selection == "4":
      print("Goodbye")
      break

    else:
      print("Enter a valid option please.")

def Balance():
  print("Your current balance is £" + str(accountBalance))
  ATMMenu()

def Deposit():
  print("Insert money")

def Withdraw():
  WithdrawMenu()

def WithdrawMenu():
  global accountBalance
  while True:
    print(" === Select withdrawal amount=== ")
    print("1. £10")
    print("2. £20")
    print("3. £50")
    print("4. £100")
    print("5. other amount")

    option = input("Select an option from 1 to 5: ")

    if option == "1" and accountBalance >= 10:
      accountBalance = accountBalance - 10
      print("Successful withdrawal")
      Balance()

    elif option == "2" and accountBalance >= 20:
      accountBalance = accountBalance - 20
      print("Successful withdrawal")
      Balance()

    elif option == "3" and accountBalance >= 50:
      accountBalance = accountBalance - 50
      print("Successful withdrawal")
      Balance()

    elif option == "4" and accountBalance >= 100:
      accountBalance = accountBalance - 100
      print("Successful withdrawal")
      Balance()

    elif option == "5":
      money = CustomWithdraw()

      accountBalance = accountBalance - money
      print("Successful withdrawal")
      Balance()


    else:
      print("Enter a valid option please.")

def CustomWithdraw():
  while True:
    withdrawAmount = input("Type the amount of money you want to withdraw: ")

    if withdrawAmount.isnumeric() == False:
      print("Enter a valid withdrawal amount please")

    elif (int(withdrawAmount))%10 != 0:
      print("Enter a valid withdrawal amount please")

    elif (int(withdrawAmount)) > accountBalance:
      print("Enter a valid withdrawal amount please")

    else:
      return int(withdrawAmount)


if __name__ == "__main__":
  PinEntry()
```
